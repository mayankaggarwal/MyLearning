```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
%matplotlib inline
```

```python
from scipy.stats import norm
```

```python
mu,sigma = 20,5
```

```python
x= norm.rvs(loc=mu,scale=sigma,size=1000)
```

```python
plt.hist(x,30,density=True)
plt.show()
```

```python
bins = np.linspace(min(x), max(x), 200)
plt.hist(x,30,density=True)
plt.plot(bins, norm.pdf(bins,loc=mu,scale=sigma),linewidth=2, color='r') #norm.pdf is complement of dnorm
plt.show()
```

# Basic Probability Distributions

For every distribution there should be four commands <br/>
1. 'd' Returns the height of distribution <br/>
2. 'p' return cumulative distribution function <br/>
3. 'q' returns inverse cummulative distribution function <br/>
4. 'r' returns randomly generated numbers

### 1. Normal (Gaussian) Distribution ()
There are four functions that can be used to generate the values associated with the normal distribution

```python
mu, sigma = 20, 5
x = np.random.normal(mu,sigma,1000)
count, bins, ignored = plt.hist(x, 30, normed=True)
normalProbabilityDensityFunction = 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) )
plt.plot(bins, normalProbabilityDensityFunction,linewidth=2, color='r')
plt.show()
```

```python
from scipy.stats import norm
mu, sigma = 20, 5
x = norm.rvs(loc=mu,scale=sigma,size=1000) #complement of rnorm
count, bins, ignored = plt.hist(x, 30, normed=True)
plt.plot(bins, norm.pdf(bins,loc=mu,scale=sigma),linewidth=2, color='r') #norm.pdf is complement of dnorm
plt.show()
```

```python
plt.plot(bins,norm.cdf(bins,loc=mu,scale=sigma),linewidth=2, color='g') #complement of pnrom
plt.show()
```

```python
norm.ppf(0.5,loc=mu,scale=sigma) #complement of qnrom
```

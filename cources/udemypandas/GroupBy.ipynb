```python
import pandas as pd
```

```python
fortune = pd.read_csv("fortune1000.csv",index_col="Rank")
fortune.head(3)
```

```python
sectors = fortune.groupby(by = "Sector") #returns a DataFrameGroupBy object
```

```python
len(sectors) #returns number of groupings same as fortune["Sector"].nunique()
```

```python
sectors.size() #retuns a series with Sectors as the index same as forture["Sector"].value_counts()
```

```python
sectors.first() #returns first row of every sector
```

```python
sectors.last() #returns last row of every sector
```

```python
sectors.groups #returns a dictionary with key as sector name and values as list of rows with that sector
```

```python
sectors.get_group("Telecommunications")
```

## Methods on the GroupBy object and DataFrame Columns

```python
sectors.max() #it looks as the left most column and returns the Company with highest alphabetical ranking
sectors.min()
```

```python
sectors.sum()
sectors.mean()
```

```python
sectors["Revenue"].sum()
sectors["Employees"].sum()
sectors[["Revenue","Profits"]].sum()
```

### Grouping by Multiple Columns

```python
sectors_industry = fortune.groupby(by=["Sector","Industry"]) #returns multi index GroupBy Dataframe
```

```python
sectors_industry.sum()
```

## The .agg() method

```python
sectors.agg({"Revenue":"sum",
             "Profits":"sum",
             "Employees":"mean"})
```

```python
sectors.agg(["size","sum","mean"]) #multipe calculations with single command
```

```python
sectors.agg({"Revenue":["sum","mean"],
             "Profits":"sum",
             "Employees":"mean"})
```

## Iterating over group by

```python
df = pd.DataFrame(columns=fortune.columns)
```

```python
for sector,data in sectors:
    highest_revenue_company_in_group = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_company_in_group) #its not inplace so with have to assign it back
```

```python
sectors["Revenue"].max()
```

```python
df
```

```python
cities = fortune.groupby("Location")
df = pd.DataFrame(columns=fortune.columns)
for city,data in cities:
    highest_revenue_company_in_city = data.nlargest(1,"Revenue")
    df = df.append(highest_revenue_company_in_city) #its not inplace so with have to assign it back
```

```python
#Importing packages
import os
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 15, 6

%matplotlib inline
```

## Checking Stationarity

#### 1. Plotting Rolling Statistics
We can plot moving average or moving variance and see if it changes over time

#### 2. Dickey Fuller Test
Here the null hypothesis is that TS is non stationary. 
The test results comprise of Test statistic and some critical values for different confidence interval.
If Test statistic is less than critical value, we can reject the null hypothesis

```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(timeseries):
    #Determine rolling statistics
    rolmean = pd.rolling_mean(timeseries,window=12) #Plotting moving avrages
    rolstd = pd.rolling_std(timeseries,window=12)
    
    #Plotting rolling statistics
    plt.plot(timeseries,color='blue',label='Original')
    plt.plot(rolmean,color="red",label='Rolling mean')
    plt.plot(rolstd,color='black',label="Rolling Std")
    plt.legend(loc='best')
    plt.title("Rolling Mean & Standard deviation")
    plt.show(block=False)
    
    #Perform Dickey-Fuller test
    print('Results of Dickey-Fuller Test')
    dftest = adfuller(timeseries['#Passengers'],autolag='AIC')
    dfoutput = pd.Series(dftest[0:4],index=['Test Statistic','p-value','#Lags used','Number of observations used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)
```

```python
test_stationarity(airpassengers)
```

## Making Time Series Stationary
There are 2 major reasons behind non-stationarity of TS
1. Trend : varying mean over time
2. Seasonality : variations at specific time frame

#### Estimating and eliminating trend
1. Using transformation like log, square root, cube etc. the trend can be removed
2. Moving average : Taking average of k consecutive values depending on frequency of time series
3. Exponentially weighted moving average : Here weights are assigned to all previous values with a decay factor

```python
#Using Transformation
fig,axs = plt.subplots(1,2)
airpassengers.plot(color='black',ax=axs[0],figsize=(12,6))
ts_log = np.log(airpassengers)
ts_log.plot(ax=axs[1])
```

```python
#Using Moving average
moving_avg = pd.rolling_mean(ts_log,window=12)
plt.plot(ts_log)
plt.plot(moving_avg,color='red')ts_log_moving_avg_diff = ts_log - moving_avg
ts_log_moving_avg_diff.dropna(inplace=True)
test_stationarity(ts_log_moving_avg_diff)
```

```python
ts_log_moving_avg_diff = ts_log - moving_avg
ts_log_moving_avg_diff.dropna(inplace=True)
test_stationarity(ts_log_moving_avg_diff)
```

```python
# Exponentially weighted moving average
expwighted_avg = pd.ewma(ts_log, halflife=12)
plt.plot(ts_log)
plt.plot(expwighted_avg, color='red')
```

```python
ts_log_ewma_diff = ts_log - expwighted_avg
test_stationarity(ts_log_ewma_diff)
```

#### Eliminating trend and seasonality
1. Differencing : In this we take the difference of observations at a particular instant with that at the preious instant
    The number of differences is called order.
2. Decomposing : In this both trend and seasonality are modeled seperately and the remaining part of the series is returned.

```python
#Differencing
ts_log_diff = ts_log - ts_log.shift()
fig,axs = plt.subplots(1,2)
ts_log.plot(color="black",label="Original",figsize=(10,4),ax=axs[0])
ts_log_diff.plot(label="Difference",figsize=(10,4),ax=axs[1])
```

```python
ts_log_diff.dropna(inplace=True)
test_stationarity(ts_log_diff)
```

```python
#Decomposing
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
decompose = seasonal_decompose(ts_log)
trend = decompose.trend
seasonal = decompose.seasonal
residual = decompose.resid

plt.subplot(411)
plt.plot(ts_log,label='Original')
plt.legend(loc='best')
plt.subplot(412)
plt.plot(trend,label="Trend")
plt.legend(loc='best')
plt.subplot(413)
plt.plot(seasonal,label="Seasonal")
plt.legend(loc='best')
plt.subplot(414)
plt.plot(residual,label="Residual")
plt.legend(loc='best')
```

```python
ts_log_decompose = residual
ts_log_decompose.dropna(inplace=True)
test_stationarity(ts_log_decompose)
```

## Forecasting a Time Series
A series with significant dependence among values. 
In this case we need to use some statistical models like ARIMA to forecast the value

### ARIMA (Auto Regressive Integrated Moving Averages)
It is a linear (like linear regression) equation in which the predictors depends on parameters p,d,q
1. Number of AR terms (p): These are lags of dependent variable.
2. Number of MA terms (q): These are lags forecast errors in prediction equation.
3. Number of differences (d) : These are the number of non seasoned differences

TO determine p,q we have use ACF and PACF

##### ACF (Auto Correlation Function)
It is the measure of correlation between TS with a lagged version of itself

```python
#ACF and PACF plots
from statsmodels.tsa.stattools import acf,pacf
```

```python
lag_acf = acf(ts_log_diff,nlags = 20)
lag_pacf = pacf(ts_log_diff,nlags = 20,method='ols')
```

```python
#Plot ACF
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title("Autocorrelation Function")
```

##### PACF (Partial Autocorrelation function)
This measures correlation between TS with lagged verion of itself 
but after eliminating the variation already explained by intervening comparisions

```python
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle="--",color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title("Partial Autocorrelation Function")
plt.tight_layout()
```

p = the log value where PACF chart crosses upper confidence value for the first time <br/>
q = the log value where ACF char crosses upper confidence value for the first time

### ARIMA Model
The p,d,q values can be specified using the order argument of ARIMA which take a tuple (p,d,q)

```python
from statsmodels.tsa.arima_model import ARIMA
```

###### AR Model

```python
model = ARIMA(ts_log, order=(2, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ts_log_diff)**2))
```

###### MA Model

```python
model = ARIMA(ts_log, order=(0, 1, 2))  
results_MA = model.fit(disp=-1)  
plt.plot(ts_log_diff)
plt.plot(results_MA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ts_log_diff)**2))
```

###### Combined Model

```python
model = ARIMA(ts_log,order=(2,1,2))
results_ARIMA = model.fit(disp=-1)
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts_log_diff)**2))
```

###### Taking it back to original

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
predictions_ARIMA_diff.head()
```

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
predictions_ARIMA_diff_cumsum.head()
```

```python
predictions_ARIMA_log = pd.Series(4.718499, index=ts_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(airpassengers)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-airpassengers)**2)/len(airpassengers)))
```

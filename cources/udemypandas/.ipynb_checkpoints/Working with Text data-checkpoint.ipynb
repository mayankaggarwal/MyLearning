```python
import pandas as pd
import numpy as np
```

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
```

```python
#chicago["Employee Annual Salary"] = [str(salary).replace('$','') for salary in chicago["Employee Annual Salary"]]
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace('$','').astype(np.float32)
```

```python
chicago["FirstName"]= [str(name).split(',')[0].strip() for name in chicago["Name"]]
chicago["LastName"]= [str(name).split(',')[1].strip() for name in chicago["Name"]]
chicago.drop("Name",axis=1,inplace=True)
```

## Common string methods - .lower(), .upper(), .title() and .len()

```python
print("HELLO World".lower())
print("HELLO World".upper())
print("HELLO World".title())
print(len("HELLO World"))
```

```python
chicago["FirstName"] = chicago["FirstName"].str.title()
```

```python
colNames = ['Position Title', 'Department', 'FirstName','LastName']
for colName in colNames:
    chicago[colName] = chicago[colName].str.title()
```

## The .str.replace() Method

```python
chicago["Department"] = chicago["Department"].str.replace("Mgmnt","Management")
```

## Filtering with String Methods

```python
mask = chicago["Position Title"].str.lower().str.contains("water")
chicago[mask]
```

```python
mask1 = chicago["Position Title"].str.lower().str.startswith("water")
mask2 = chicago["Position Title"].str.lower().str.endswith("ist")
chicago[mask1 & mask2]
```

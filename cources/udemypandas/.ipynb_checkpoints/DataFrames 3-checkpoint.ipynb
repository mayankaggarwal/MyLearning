## Retrieve rows by Index Label with .loc[]

```python
bond.reset_index(inplace=True)
bond.set_index("Film",inplace=True)
bond.sort_index(inplace=True)
bond.head(3)
```

```python
bond.loc["Goldfinger"] #it returns the Series where column names becomes index and row values becomes Series values
```

```python
bond.loc["Casino Royale"]
bond.loc["GolderEye" : "SkyFall"]
bond.loc[["Octopussy","Moonraker"]] #if any index key does not exist than it does not throughs error and returns row with NaN
```

```python
bond.iloc[15:21]
bond.iloc[:4]
bond.iloc[[1,5,20]]
```

```python
print("ix[] is combination of loc and iloc")
bond.ix["Casino Royale"]
bond.ix[5:10]
# bond.ix[100] 
print("if no value at index position than error is thrown  but if value doesn't exist than NaN occurs")
```

```python
#Selecting specific columns using second paramter
bond.loc["Moonraker",["Year","Actor"]]
```

```python
bond.iloc[1:5,[2,5]] #accepts only index of columns and rows index
```

```python
bond.ix["Casino Royale",["Actor","Director"]]
bond.ix[:"Casino Royale",4]
```

## Creating a ramdon sample

```python
bond = pd.read_csv("jamesbond.csv",index_col="Film")
bond.sort_index(inplace=True)
bond.head(2)
```

```python
bond.sample(n=5) #returns a brand new dataframe
bond.sample(frac=.25) #returns a fraction of rows to be returned
bond.sample(n=1,axis=1) #selecting ramdom columns
```

```python
bond.sort_values("Box Office",ascending=False).head(3)
```

```python
bond.nlargest(3,"Box Office") #its more optimal than sort_values 
```

```python
bond.nsmallest(3,"Box Office")
```

```python
bond["Budget"].nlargest(3) #returns a series
```

## Filtering with the where Method

## The .query() Method

```python
#it woukd only if there are not spaces in columns
bond.columns = [columnName.replace(" ","_") for columnName in bond.columns]
```

```python
bond.query("Actor == 'Sean Connery'")
```

```python
bond.query("Actor == 'Roger Moore' and Director == 'John Glen'")
```

```python
bond.query("Actor in ['Timothy Dalton','George Lazenby']")
```

## A review of .apply Method on Single Column

```python
def convert_to_string_and_add_millions(number):
    return str(number) + ' Millions!'
```

```python
bond["Box_Office"] = bond["Box_Office"].apply(convert_to_string_and_add_millions)
```

```python
columns = ["Budget","Bond_Actor_Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

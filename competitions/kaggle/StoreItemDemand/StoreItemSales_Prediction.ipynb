```python
train = data.loc[data['sales'].notnull(),['date','store','item','sales']]
```

```python
store1item1 = train.loc[(train['store']==1) & (train['item']==1),['date','sales']]
store1item1['sales'] = store1item1['sales'].astype(np.int16)
store1item1.set_index(['date'],inplace=True)
```

```python
plt.figure(figsize=(16,8))
plt.plot(store1item1)
plt.show()
```

```python
ts_log = np.log(store1item1)
```

```python
ts_log_diff = ts_log - ts_log.shift()
plt.figure(figsize=(16,8))
plt.plot(ts_log_diff)
plt.show()
```

```python
ts_log_diff.dropna(inplace=True)
```

```python
from statsmodels.tsa.stattools import acf,pacf 
```

```python
lag_acf = acf(ts_log_diff,nlags=20)
```

```python
lag_pacf = pacf(ts_log_diff,nlags=20,method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(ts_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```

From above p and q both looks to be 1

```python
from statsmodels.tsa.arima_model import ARIMA
```

```python
model = ARIMA(ts_log, order=(1, 1, 0))  
results_AR = model.fit(disp=-1)  
plt.figure(figsize=(16,8))
plt.plot(ts_log_diff)
plt.plot(results_AR.fittedvalues, color='red')
#plt.title('RSS: %.4f'% sum((results_AR.fittedvalues-ts_log_diff)**2))
```

```python
model = ARIMA(ts_log, order=(0, 1, 2))  
results_MA = model.fit(disp=-1) 
plt.figure(figsize=(16,8))
plt.plot(ts_log_diff)
plt.plot(results_MA.fittedvalues, color='red')
#plt.title('RSS: %.4f'% sum((results_MA.fittedvalues-ts_log_diff)**2))
```

```python
model = ARIMA(ts_log, order=(2, 1, 2))  
results_ARIMA = model.fit(disp=-1)  
plt.figure(figsize=(16,8))
plt.plot(ts_log_diff)
plt.plot(results_ARIMA.fittedvalues, color='red')
#plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts_log_diff)**2))
```

```python
predictions_ARIMA_diff = pd.Series(results_ARIMA.fittedvalues, copy=True)
print(predictions_ARIMA_diff.head())
```

```python
predictions_ARIMA_diff_cumsum = predictions_ARIMA_diff.cumsum()
print(predictions_ARIMA_diff_cumsum.head())
```

```python
predictions_ARIMA_log = pd.Series(ts_log.ix[0], index=ts_log.index)
predictions_ARIMA_log = predictions_ARIMA_log.add(predictions_ARIMA_diff_cumsum,fill_value=0)
predictions_ARIMA_log.head()
```

```python
predictions_ARIMA = np.exp(predictions_ARIMA_log)
plt.plot(store1item1)
plt.plot(predictions_ARIMA)
plt.title('RMSE: %.4f'% np.sqrt(sum((predictions_ARIMA-store1item1)**2)/len(store1item1)))
```

```python
p = d = q = range(0,2)
```

```python
import itertools
pdq = list(itertools.product(p,d,q))
```

```python
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

print('Examples of parameter combinations for Seasonal ARIMA...')
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[1]))
print('SARIMAX: {} x {}'.format(pdq[1], seasonal_pdq[2]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[3]))
print('SARIMAX: {} x {}'.format(pdq[2], seasonal_pdq[4]))
```

```python
import warnings;
import statsmodels.api as sm
warnings.filterwarnings("ignore") # specify to ignore warning messages

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = sm.tsa.statespace.SARIMAX(store1item1,
                                            order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)

            results = mod.fit()

            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
        except:
            print('Exception')
            continue
```
